# Gráfica de ejemplo - realizaciones de viento

Las series de tiempo mostradas corresponden a una velocidad de viento **$U_W$=11.4 m/s**, y semillas **$S_1,S_{14},S_{24}$** y **$S_{71}$**

## 1. Importando librerías

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## 2. Importando datos

```python
wndTS = pd.read_excel(r'..\data\wind_realizations.xlsx')
```

```python
wndTS.head()
```

## 3. Estableciendo variables de control del gráfico 

```python
ps_w, ps_h = 15,4.5 #Dimensiones de cada gráfico indvidual
line_w = 0.8      #ancho de línea
axFont = 20       #Tamaño del texto en eje
num_axFont = 16   #Tamaño del números en el eje

grid_dash = (8, 4) #Controla la escala de discontinuidad del grid
grid_lw = 0.5      #ancho de línea del grid

plot_col = ['black', '#3C4A85', '#CE4636', '#22B14C']  #paleta de colores de los gráficos

n_plots = wndTS.shape[1]-1 #Número de gráficos
```

```python
p_title = ['Velocidad de viento - $U_W$ = 11.4 m/s y $S_{1}$', 
           'Velocidad de viento - $U_W$ = 11.4 m/s y $S_{14}$',
           'Velocidad de viento - $U_W$ = 11.4 m/s y $S_{24}$',
           'Velocidad de viento - $U_W$ = 11.4 m/s y $S_{71}$']
```

## 4. Creando gráficos de viento 

```python
#Creando figura y subplots
fig, axs = plt.subplots(n_plots, 1, figsize = (ps_w, n_plots*ps_h))

for i in range(n_plots):
    axs[i].plot(wndTS['Tiempo [s]'], wndTS[wndTS.columns[i+1]],
               linewidth = line_w,
               color = plot_col[i])
    
    axs[i].set_title(p_title[i], fontsize = axFont + 2)
    axs[i].set_ylabel('$V_X$ [m/s]', fontsize = axFont)
    
    axs[i].grid(color = 'gray', linestyle = '--', linewidth = grid_lw, alpha = 0.5, dashes = grid_dash)
    
    axs[i].tick_params(axis='both', labelsize = num_axFont)
    axs[i].set_xlim(wndTS.iloc[0, 0], wndTS.iloc[-1, 0])
    axs[i].set_ylim(0, 20)
    
axs[n_plots-1].set_xlabel('Tiempo [s]', fontsize = axFont)
fig.tight_layout()
plt.show()   
```
