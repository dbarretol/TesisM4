# Gráfica de ejemplo - realizaciones de diversas respuestas dinamicas

Las series de tiempo mostradas corresponden a una velocidad de viento $U_W$=11.4 m/s y semilla $S_{50}$

## 1. Importando librerías

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## 2. Importando datos

```python
DR = pd.read_excel(r'..\data\response_realizations (Uw11.4_S50).xlsx')
```

```python
DR.head()
```

## 3. Estableciendo variables de control de gráfico

```python
ps_w, ps_h = 15,3.5 #Dimensiones de cada gráfico indvidual
line_w = 0.8      #ancho de línea
axFont = 20       #Tamaño del texto en eje
num_axFont = 16   #Tamaño del números en el eje

grid_dash = (8, 4) #Controla la escala de discontinuidad del grid
grid_lw = 0.5      #ancho de línea del grid

plot_col = ['black', '#3C4A85', '#CE4636', '#22B14C']  #paleta de colores de los gráficos
```

```python
#Estableciendo los titulos de las figuras individuales
p1_title = ['Desplazamiento de la punta de la torre - $U_W$ = 11.4 m/s y $S_{50}$',
           'Fuerza de corte en la base de la torre - $U_W$ = 11.4 m/s y $S_{50}$',
           'Momento flector en la base de la torre - $U_W$ = 11.4 m/s y $S_{50}$']
p2_title = ['Desplazamiento de la punta del álabe - $U_W$ = 11.4 m/s y $S_{50}$',
           'Fuerza de corte en la raíz del álabe - $U_W$ = 11.4 m/s y $S_{50}$',
           'Momento flector en la raíz del álabe - $U_W$ = 11.4 m/s y $S_{50}$']
p3_title = ['Potencia eléctrica generada - $U_W$ = 11.4 m/s y $S_{50}$',
           'Torque del generador - $U_W$ = 11.4 m/s y $S_{50}$',
           'Velocidad rotacional del generador - $U_W$ = 11.4 m/s y $S_{50}$']
p4_title = ['Ángulo del álabe - $U_W$ = 11.4 m/s y $S_{50}$',
           'Torque del rotor - $U_W$ = 11.4 m/s y $S_{50}$',
           'Velocidad rotacional del rotor - $U_W$ = 11.4 m/s y $S_{50}$']
```

```python
#Estableciendo las columnas de los datos a graficar por cada grupo
indG1 = [0,6, 10, 11]
indG2 = [0, 2, 7, 8]
indG3 = [0, 12, 13, 5]
indG4 = [0, 3, 9, 4]
```

### 3.1. Graficando primer grupo

```python
DR_G1 = DR.iloc[:,indG1]
```

```python
DR_G1.head()
```

```python
#Creando figuras y subplots
n_plots = DR_G1.shape[1]-1

fig, axs = plt.subplots(n_plots, 1, figsize = (ps_w, n_plots*ps_h))

for i in range(n_plots):
    axs[i].plot(DR_G1['Time [s]'], DR_G1[DR_G1.columns[i+1]],
               linewidth = line_w,
               color = plot_col[i])
    
    axs[i].set_title(p1_title[i], fontsize = axFont + 2)
    axs[i].set_ylabel(DR_G1.columns[i+1], fontsize = axFont)
    
    axs[i].grid(color = 'gray', linestyle = '--', linewidth = grid_lw, alpha = 0.5, dashes = grid_dash)
    
    axs[i].tick_params(axis='both', labelsize = num_axFont)
    axs[i].set_xlim(DR_G1.iloc[0, 0], DR_G1.iloc[-1, 0])
    
axs[n_plots-1].set_xlabel('Tiempo [s]', fontsize = axFont)
fig.tight_layout()
plt.show()   
```
